```python
### Saving all the jacket(s), t-shirt(s), and shirt(s)  that are currently present on Supreme 

# Rendering the HTML Shop page 
html = browser.html
page_soup = soup(html, 'html.parser')

jackets = page_soup.find_all("li",class_ = "jackets")
tshirts = page_soup.find_all("li", class_ = "t-shirts")
shirts = page_soup.find_all("li",class_ = "shirts")

jacketList = []
tshirtList = []
shirtList= []

for i in range(0,len(jackets)):
    jacketList.append(jackets[i].a['href'])
    print(jackets[i].a['href'])
    
for i in range(0,len(tshirts)):
    tshirtList.append(tshirts[i].a['href'])
    print(tshirts[i].a['href'])
    
for i in range(0,len(shirts)):
    shirtList.append(shirts[i].a['href'])
    print(shirts[i].a['href'])
```

```python
# Rendering the HTML Shop page 
html = browser.html
page_soup = soup(html, 'html.parser')

## Setting up the HREF for the jacket
shirtHref = page_soup.find_all("li", class_ = "t-shirts")[2].a['href']

browser.click_link_by_href(shirtHref)

### On the "Add to cart page"

## Clicking sizes for a Large 
browser.find_by_text('Medium').click()

# Heather grey color selection by Xpath
heatherGrey = "//button[@data-style-name = 'Heather Grey']"
browser.find_by_xpath(heatherGrey).click()

time.sleep(1.5)
## Clicking "Add to cart" 
browser.find_by_value("add to cart").click()


time.sleep(1.5)
## Clicking the Check out button
browser.find_by_text('checkout now').click()

### ON THE PAYMENT PAGE###

## Filling out the name field with my full-name 
browser.find_by_id("order_billing_name").fill("Carlos Jaramillo")


## Filling out the email field with my email 
browser.find_by_id("order_email").fill("Carlosr03@gmail.com")


## Filling out the telephone field with my phone number 
browser.find_by_id("order_tel").fill("5146545127")


## Filling out my address for the billing address
browser.find_by_name("order[billing_address]").fill("200 Woolner Avenue")


## Filling in the apartment number for myself 
browser.find_by_name("order[billing_address_2]").fill("207")


## Clicking the CANADA button 
browser.find_by_value("CANADA").click()


## Filling out the "Zip code" field with my zip code 
browser.find_by_id("order_billing_zip").fill("m6n1y4")


## Filling out the city field with Toronto
browser.find_by_id("order_billing_city").fill("Toronto")


## Filling out the province option
browser.find_by_value("ON").click()


## Filling out the credit card information
browser.find_by_id("rnsnckrn").fill("1111111111111")


## picking the expiry month for the credit card
browser.find_by_value("11").click()


## picking the expiry year for the credit card
browser.find_by_value("2020").click()


## picking the CVV for the credit card
browser.find_by_id("orcer").fill("111")
```
